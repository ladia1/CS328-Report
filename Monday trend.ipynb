```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

start_date = pd.to_datetime('2020-05-31')
end_date = pd.to_datetime('2020-06-30')

# Load data
cases = pd.read_csv(r"C:\Users\Dishant\Desktop\Untitled Folder\case_time_series.csv")
cases["Date"] = pd.to_datetime(cases["Date_YMD"], format='%Y-%m-%d')
cases.set_index("Date", inplace=True)
cases = cases["Daily Confirmed"]
cases_train = cases.loc[start_date:end_date]

# Define moving average window size
window_size = 7

# Calculate moving average
moving_avg = cases_train.rolling(window_size).mean()

# Define figure size and font size
fig, ax = plt.subplots(figsize=(10,6))
plt.rcParams.update({'font.size': 12})

# Plot data and moving average
ax.plot(cases_train.index, cases_train)

# Add labels, legend, and title
ax.set_xlabel('Date June 2020')
ax.set_ylabel('No. of cases')
ax.set_title('Trend of daily new cases in India')

# Set x-axis ticks and labels
ax.xaxis.set_major_locator(mdates.DayLocator(interval=1))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=7))
ax.tick_params(axis='x', which='minor', bottom=False, labelbottom=False)
for tick in ax.xaxis.get_minor_ticks():
    tick.tick1line.set_visible(False)
    tick.tick2line.set_visible(False)
    tick.label1.set_visible(False)
plt.xticks(rotation=90)

# Add vertical dotted lines for Mondays
mondays = pd.date_range(start=start_date, end=end_date, freq='W-MON')
for monday in mondays:
    ax.axvline(x=monday, linestyle='--', color='grey', alpha=0.5)

plt.show()
```
